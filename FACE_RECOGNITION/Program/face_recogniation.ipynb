```python
import os
import cv2
import numpy as np

#KNN-ALGO

def dist(X1,X2):
    return np.sqrt(sum((X2-X1)**2))
    
def KNN(pixels,Labels,X):
    
    m=pixels.shape[0]
    
    expected_value=[]
    
    for i in range(m):
        d=dist(pixels[i],X)
        expected_value.append((d,Labels[i]))
        
    expected_value=sorted(expected_value)
    
    expected_value=expected_value[:20]
    
    expected_value=np.array(expected_value)
    
    unique=np.unique(expected_value[:,1],return_counts=True)
    
    
    ind =unique[1].argmax() 
    
    pre=unique[0][ind]
    
    return pre


# preparation of traing data from system
data_path="/home/manik/data/"
data_train=[]
labels=[]
names={}
class_id=0
for file in os.listdir(data_path):
    if(file.endswith('.npy')):
        names[class_id]=file[:-14]
        fg=np.load(data_path+file)
        data_train.append(fg)
        
        labels.append(class_id*np.ones(fg.shape[0],))
        class_id+=1
        
data_train=np.concatenate(data_train,axis=0)

labels=np.concatenate(labels,axis=0).reshape(-1,1)

train_data_set=np.concatenate((data_train,labels),axis=1)

#Testing

cap=cv2.VideoCapture(0)

face_cascade=cv2.CascadeClassifier("/home/manik/haarcascade_frontalface_alt.xml")

if face_cascade.empty():
    print(True)

while True:
    
    ret,frame=cap.read()
    
    # gray_frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    if ret==False:
        continue
        
    faces = face_cascade.detectMultiScale(frame,1.3,5)
    
    if len(faces)==0:
        continue
        
    faces= sorted(faces,key=lambda f:f[2]*f[3])
    
    for (x,y,w,h) in faces[-1:]:
                
        face_section=frame[y-10:y+h+10,x-10:x+w+10]
        
        face_section=cv2.resize(face_section,(100,100))
        
        output=KNN(train_data_set[:,:-1],train_data_set[:,-1],face_section.flatten())
        
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,255),2)

        face_name=names[int(output)]

        print(face_name)
        
        cv2.putText(frame,face_name,(x,y-10),cv2.FONT_,1,(255,0,0),cv2.LINE_AA)
        
    cv2.imshow("Video Frame",frame)
    
    key_pressed = cv2.waitKey(1) & 0xFF
    
    if(key_pressed==ord('q')):
        break

cap.release()

cv2.destroyAllWindows()
```
