```python
import cv2

# inside videocapture id of webcam come if we have multiple webcam otherwise 0 for default web cam

cap=cv2.VideoCapture(0)

face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
if face_cascade.empty():
    print(True)

while True:
    
    # cap.read() return 2 value 1 is bool whether it find webcam or not and is frame of video 
    ret,frame=cap.read()
    
    gray_frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    if ret==False:
        continue
        
    faces = face_cascade.detectMultiScale(gray_frame,1.3,5)
    
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        
    # shows the frame returned by cap.read()
    cv2.imshow("Video Frame",frame)
    #cv2.imshow("gray frame",gray_frame)
    
    # waitkey is normally used for how much time we want video visible to user but here we are using to take input key and converting 32
    # bit to 8 bit by multipling 0xFF
    key_pressed = cv2.waitKey(1) & 0xFF
    
    if(key_pressed==ord('q')):
        break
        
cap.release()

cv2.destroyAllWindows()
    
```
